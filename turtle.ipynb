```python
from ColabTurtle.Turtle import *
from math import sqrt
initializeTurtle(initial_speed=10)
color("red")
pensize(1)
bgcolor("yellow")
left(90)
for i in range(2000):
  forward(20)
  k= i%90
  left(math.sqrt(i))
  left(k)

hideturtle()
```
